```python
#%matplotlib qt
%matplotlib inline

import numpy
import matplotlib.pyplot as plt
```

```python
def makeStabilityRegionPlot(g,d,c,m,largeC,largeP,p,axes,
                            title="Stability Region For The ODEs",
                            xLabelOffsetx=0.95,
                            xLabelOffsety=-0.02):
    
    #axes.set_xlabel(r"$p=1+m\theta$", fontsize=25)
    #axes.set_ylabel(r"$u=\frac{cd}{g-d}$",labelpad=-20, fontsize=25)
    axes.annotate(r"$p$", 
                  xy=(xLabelOffsetx,xLabelOffsety), xytext=(xLabelOffsetx,xLabelOffsety),
                  ha='left', va='top', fontsize=25,
                  xycoords='axes fraction', textcoords='offset points')
    axes.annotate(r"$c$", #"$u=\frac{cd}{g-d}$", 
                  xy=(-0.07,0.95),xytext=(-0.07,0.95),
                  ha='left', va='top', fontsize=25, rotation=90,
                  xycoords='axes fraction', textcoords='offset points')
    axes.set_title(title, fontsize=35)
    #axes.ticklabel_format(axis='y', style='sci', scilimits=(-1,1))
    
    axes.set_xticks(numpy.array([1.0,1.0+m])) 
    axes.set_xticklabels(["1","1+m"])

    axes.set_yticks(numpy.array([(g-d)/(g+d),
                                 #(g-d)/d,
                                 #(g-d)/(g+d)*(1+m),
                                 (g-d)/d*(1+m)])) 
    axes.set_yticklabels([r"$\frac{g-d}{g+d}$",
                          #r"$\frac{g-d}{d}$",
                          #r"$\frac{g-d}{g+d}(1+m)$",
                          r"$\frac{g-d}{d}(1+m)$"])
    axes.tick_params(axis='both', which='major', labelsize=25)
    axes.tick_params(axis='both', which='minor', labelsize=30)
    
    
    pShade       = numpy.array([1.0,1+m])
    pShadeTop    = numpy.array([(g-d)/d,(g-d)/d*(1+m)])
    pShadeBottom = numpy.array([(g-d)/(g+d),(g-d)/(g+d)*(1+m)])
    
    axes.fill_between(pShade,pShadeTop,pShadeBottom,facecolor="grey")
    axes.plot(p,(g-d)/(g+d)*p,'r',linewidth=2)
    axes.plot(p,(g-d)/d*p,'r',linewidth=2)
    axes.plot(numpy.array([1.0,1.0]),numpy.array([0.0,(g-d)/d]),'k:',linewidth=2)
    axes.plot(numpy.array([1.0+m,1.0+m]),numpy.array([0.0,(g-d)/d*(1.0+m)]),'k:',linewidth=2)
    
    axes.text(largeP-0.7,(g-d)/d*(largeP-0.7),r"$c=\frac{g-d}{d}p$", 
              horizontalalignment='right', verticalalignment='bottom',fontsize=25)
    axes.text(largeP-0.5,(g-d)/(g+d)*(largeP-0.5),r"$c=\frac{g-d}{g+d}p$", 
              horizontalalignment='left', verticalalignment='top', fontsize=25)
    
    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    axes.spines['left'].set_position('zero')
    axes.spines['bottom'].set_position('zero')
    
    # Eliminate upper and right axes
    axes.spines['right'].set_color('none')
    axes.spines['top'].set_color('none')
    #axes.spines['left'].set_color('none')
    #axes.spines['bottom'].set_color('none')
    
    
    # Show ticks in the left and lower axes only
    axes.xaxis.set_ticks_position('bottom')
    axes.yaxis.set_ticks_position('left')
    
    
    axes.set_xlim(-0.1, largeP)
    axes.set_ylim(-0.1, largeC)
```
